```python
data2020 = pd.read_csv('data/원본/서울특별시_부동산_실거래가_정보_2020년.csv')
```

```python
data2020
```

```python
cols = ['자치구명', '신고년도', '건물면적', '층정보', '건물주용도', '물건금액', '건축년도']

data2020 = data2020[cols]
data_apt = data2020[data2020['건물주용도']=='아파트']
data_apt
```

```python
df = pd.DataFrame(columns=cols)
df = pd.concat([df, data_apt], axis=0)
df.set_index('자치구명', inplace=True)
df.reset_index(inplace=True)
```

```python
df
df.to_csv('data/서울특별시_아파트_실거래가_정보_2020년.csv')
```

```python
cols = ['자치구명', '신고년도', '건물면적', '층정보', '건물주용도', '물건금액', '건축년도']

for i in range(len(data)):
    data[i] = data[i][cols]
```

```python
df = pd.DataFrame(columns = cols)
for i in range(len(data)):
    df = pd.concat([df, data[i]], axis=0)
```

```python
df.set_index('자치구명', inplace=True)
df.reset_index(inplace=True)
```
